```python
import pandas as pd
import numpy as np
```

```python
ipl = pd.read_csv('ipl_dataset.csv')
#print(ipl)
```

```python
#Dropping certain features
df = ipl.drop(['full_scorecard','margin', 'man_of_the_match', 'stadium', 'place'], axis =1)
```

```python
df.head(200)
```

```python
df.groupby('winner').describe()
```

```python
# prompt: drop Deccan Chargers , Delhi Daredevils	 , Gujarat Lions , Kochi Tuskers Kerala , Rising Pune Supergiant	, Rising Pune Supergiants from the entire dataset

df = df.drop(df.loc[df['winner'].isin(['Match','No','Tied','Chargers', 'Daredevils', 'Gujarat Lions', 'Kochi', 'Supergiant', 'Supergiants','Pune Warriors'])].index)
```

```python
df.groupby('winner').describe()
```

```python
toss_winners3 = {}
for i, row in df.iterrows():
  if row['toss_winner'] not in toss_winners3:
    toss_winners3[row['toss_winner']] = len(toss_winners3)
```

```python
#df.to_dict('dict')
teams = {"Kings": 0, "Mumbai": 1, "Royals" : 2, "Super" : 3, "Royals" : 4, "Sunrises"  : 5, "Capitals" : 6,  "Punjab"  : 7,  "KKR" : 8, "Punjab" : 9, "Titans" : 10 ,"Guj" :11}
```

```python
df['winner1'] = df['winner'].apply(lambda x:teams.get(x,-1))
df['new_toss'] = df['toss_winner'].apply(lambda x:toss_winners3.get(x,-1))
dic = {'Field': 0 ,'Bat':1}
df['newc'] = df['toss_choice'].apply(lambda x:dic.get(x,-1))
```

```python
df.head()
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
x= df[['new_toss','newc']]
y= df['winner1']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```
